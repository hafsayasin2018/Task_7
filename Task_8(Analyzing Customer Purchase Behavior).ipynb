```python
import pandas as pd
import os
import numpy as np
```

```python
#path = "./Sales_Data"
data_set = pd.read_excel("online_retail_II.xlsx")
data_set
```

### 1.Cleaning and preprocessing the data to remove duplicates, missing values, and other errors.

```python
#Remove duplicates:
print(data_set.duplicated())
```

```python
remove_duplicate = data_set.drop_duplicates(inplace = False)
remove_duplicate
```

```python
# Remove missing values
remove_duplicate.isna().sum() # count missing values
```

```python
remove_missing = remove_duplicate.dropna()
print(remove_missing)
```

```python
remove_missing.info()
```

```python
print((remove_missing['Quantity'] < 0).any())
print((remove_missing['Price'] < 0).any())
print((remove_missing['Customer ID'] < 0).any())
```

```python
# remove negative values
remove_neg_values = remove_missing[(remove_missing['Price'] > 0) & (remove_missing['Quantity']>0)]
remove_neg_values
```

```python
print((remove_neg_values['Quantity'] < 0).any())
print((remove_neg_values['Price'] < 0).any())
```

```python
# check if any zero values
print((remove_neg_values['Quantity'] == 0).any())
print((remove_neg_values['Price'] == 0).any())
print((remove_neg_values['Customer ID'] == 0).any())
```

### 2.Merging and aggregating the data to create summary tables and metrics, such as total sales by product or customer.

```python
remove_neg_values['Sales'] = remove_neg_values['Quantity'] * remove_neg_values['Price']
new_dataset = remove_neg_values
new_dataset
```

### 3.Analyzing the data to identify trends and patterns in customer purchase behavior, such as which products are frequently purchased together, which products are popular at different times of year, or which customers are most valuable to the business.

```python
new_dataset['months']=new_dataset['InvoiceDate'].astype(str).str[5:7]
new_dataset['year']=new_dataset['InvoiceDate'].astype(str).str[0:4]
new_dataset['time']=new_dataset['InvoiceDate'].astype(str).str[10:17]
new_dataset
```
